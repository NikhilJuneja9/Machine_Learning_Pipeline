```python
import pandas as pd
import numpy as np
import os
from IPython import  get_ipython
from sklearn.model_selection import train_test_split
from sklearn import pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
#making dataset path relative
root_dir = os.path.dirname(os.path.abspath(get_ipython().starting_dir))
Dataset_path = os.path.join(root_dir,"Dataset","titanic_dataset.csv")
```

```python
df = pd.read_csv(Dataset_path)
df.sample(5)
```

```python
df.drop(columns = ["Name","Ticket","Cabin"],inplace=True)
```

```python
df.set_index(["PassengerId"],inplace=True)
```

```python
x_test,x_train,y_test,y_train = train_test_split(df.drop(columns=["Survived"]),df["Survived"],random_state=42,train_size=0.20)
x_test
```

```python
trf1 = ColumnTransformer([("impute_age",SimpleImputer(),[2]),("impute_embarked",SimpleImputer(strategy="most_frequent"),[6])],remainder="passthrough")
```

```python
trf2 = ColumnTransformer([("encoded_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[6]),("sex_encoded",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1])],remainder="passthrough")
```

```python
trf3 = ColumnTransformer([("scaling",MinMaxScaler(),slice(0,10))])
```

```python
trf4 = DecisionTreeClassifier()
```

```python
#making pipeline

pipe = make_pipeline(trf1,trf2,trf3,trf4)
```

```python
pipe.fit(x_train,y_train)
```
